Pytorch Computational Graph

```python
a = torch.tensor(2.0, requires_grad = True)
b = torch.tensor(3.0, requires_grad = True)
x = torch.tensor(4.0, requires_grad = True)

y = a + b
z = x * y
z
```

```python
print(z.grad_fn)
print(y.grad_fn)
print(a.grad_fn)
```

```python
# * vs @
# element wise multiplication (*) multiplies numbers in the same positions
# of course, the tensors must have the exact same shape

tensor_a = torch.tensor([[1, 2, 3], [4, 5, 6]])
tensor_b = torch.tensor([[1, 2, 3], [4, 5, 6]])

tensor_a * tensor_b
```

```python
# Matrix Multiplication (@)
# the number of columns in the first matrix must match the number of rows in the second matrix

tensor_a = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype = torch.float)
tensor_b = torch.tensor([[1, 2, 3], [4, 5, 6], [4, 5, 6]], dtype = torch.float)

tensor_a @ tensor_b

#When we use this formula: Z = XW + b, we will always use the @ symbol
```

Reduction Operations

```python
#sum(), mean(), min(), max()
# we can collapse with the specific dimension
# dim = 0 collapses vertically
# dim = 1 collapses horizontally

average_per_feature = tensor_b.mean(dim = 0)
average_per_row = tensor_b.mean(dim = 1)

print(average_per_feature)
print(average_per_row)
```
